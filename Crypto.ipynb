# PART 1 - ORGANIZING JSON FROM API (LIVE PRICES)

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import requests
import pprint
from functools import reduce
```

```python
api_key = pd.read_csv('coinmarketcapAPI.csv')
api_key = api_key[api_key['API'] == "COINMARKETCAP" ]['KEY'][0]
```

```python
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': api_key,
}
```

```python
#SETTING THE URL AND PARAMETERS


url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'100',
  'convert':'USD',    
}
```

```python
json_data = requests.get(url, params=parameters, headers=headers).json()
pprint.pprint(json_data)
```

```python
#AS WE CAN SEE EVERYTHING IS INSIDE DICTIONARY NAMED 'data'
# FOR OUR DATAFRAME WE NEED ONLY id, name, symbol, market_cap, price FROM data dictionary

raw_data=[]
for x in json_data['data']:
   raw_data.append((x['last_updated'],x['id'],x['name'],x['symbol'],x['quote']['USD']['market_cap'],x['circulating_supply'],x['quote']['USD']['price']))
```

```python
#Formating the Table so that large numbers can fit in
def format_float(value):
    return f'{value:,.2f}'

pd.options.display.float_format = format_float
```

```python
crypto=pd.DataFrame(raw_data, columns=['Date','id','name','Symbol','market_cap','supply', 'Price'])
```

```python
crypto["Date"] = pd.to_datetime(crypto["Date"], utc=True)
```

```python
crypto["Date"] = crypto["Date"].dt.date   
```

# PART 2 - DISCOVERING HISTORICAL DATA - DATA CLEANING

UNTILL SEPTEMBER 2021

```python
ind = pd.read_csv("index_us.csv")
index = pd.DataFrame(ind[["Date", "Close"]])#0
index["Symbol"] = "DXY"



bit = pd.read_csv("Bitcoin.csv")
bitcoin = pd.DataFrame(bit[["Date", "Price"]])#1
bitcoin["Symbol"] = "BTC"

eth = pd.read_csv("Ethereum.csv")
ethereum = pd.DataFrame(eth[["Date", "Price"]])#2
ethereum["Symbol"] = "ETH"


car = pd.read_csv("Cardano.csv")
cardano = pd.DataFrame(car[["Date", "Price"]])#3
cardano["Symbol"] = "ADA"


bina = pd.read_csv("Binance.csv")
binance = pd.DataFrame(bina[["Date", "Price"]])#4
binance["Symbol"] = "BNB"

xr = pd.read_csv("XRP.csv")
xrp = pd.DataFrame(xr[["Date", "Price"]])#5
xrp["Symbol"] = "XRP"


sol = pd.read_csv("Solana.csv")
solana = pd.DataFrame(sol[["Date", "Price"]])#6
solana["Symbol"] = "SOL"


dogecoin= pd.read_csv("Dogecoin.csv")
doge = pd.DataFrame(dogecoin[["Date", "Price"]])#7
doge["Symbol"] = "DOGE"

dot = pd.read_csv("Polkadot.csv")
polkadot = pd.DataFrame(dot[["Date", "Price"]])#8
polkadot["Symbol"] = "DOT"


uni = pd.read_csv("Uniswap.csv")
uniswap = pd.DataFrame(uni[["Date", "Price"]])#9
uniswap["Symbol"] = "UNI"


link = pd.read_csv("Chainlink.csv")
chainlink = pd.DataFrame(link[["Date", "Price"]])#10
chainlink["Symbol"] = "LINK"
```

```python
index = index.rename({"Close":"Price"}, axis=1)
```

```python
index["Date"] = pd.to_datetime(index["Date"], utc=True)
```

```python
from functools import reduce

dfs= [bitcoin,ethereum, cardano, binance, xrp, solana, doge, polkadot, uniswap, chainlink]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'], how='outer'), dfs)
```

```python
df_merged["Date"] = pd.to_datetime(df_merged["Date"], utc=True)
```

```python
df_merged = df_merged.merge(index, how="left", on="Date")
```

```python
df_merged["Date"] = df_merged["Date"].dt.date
```

```python
new_col = ['Date', 'BTC', 'Symbol_x', 'ETH', 'Symbol_y', 'ADA',
       'Symbol_x', 'BNB', 'Symbol_y', 'XRP', 'Symbol_x', 'SOL',
       'Symbol_y', 'DOGE', 'Symbol_x', 'DOT', 'Symbol_y', 'UNI',
       'Symbol_x', 'LINK', 'Symbol_y', 'DXY', 'Symbol']
```

```python
df_merged.columns = new_col
```

```python
df_merged.drop(["Symbol_x","Symbol_y","Symbol"], axis=1, inplace=True)
```

```python
df_merged['BTC'] = df_merged['BTC'].apply(lambda x: str(x).replace(",",""))
df_merged['ETH'] = df_merged['ETH'].apply(lambda x: str(x).replace(",",""))
```

```python
df_merged['BTC'] = pd.to_numeric(df_merged['BTC'])
```

```python
df_merged['ETH'] = pd.to_numeric(df_merged['ETH'])
```

```python
df_merged["Date"] = pd.to_datetime(df_merged["Date"], utc=True)
```

```python
df_merged["Date"] = df_merged["Date"].dt.date
```

# PART 4: ROI

```python
df_merged = df_merged.sort_values(["Date"])
```

```python
stocklist = ['ETH','ADA', 'BNB','XRP','SOL','DOGE','DOT','UNI', 'LINK']
```

```python
#CALCULATING RETURN ON INVESTMENTS

coin_roi = ['BTC','ETH','ADA','BNB','XRP','DOGE','LINK']
roi=[]
for s in coin_roi:
      roi.append(((df_merged[s][0] - df_merged[s][396]) / df_merged[s][396])*100)
dict_roi = dict(zip(coin_roi,roi))   

dict_roi['SOL'] = ((df_merged['SOL'][0] - df_merged['SOL'][110]) / df_merged['SOL'][110])*100
dict_roi['DOT'] = ((df_merged['DOT'][0] - df_merged['DOT'][383]) / df_merged['DOT'][383])*100
dict_roi['UNI'] = ((df_merged['UNI'][0] - df_merged['UNI'][306]) / df_merged['UNI'][306])*100
dict_roi
```

# PART 4  - DAILY CHANGE PERCENTAGE RETURN

```python
df_merged = df_merged.sort_values(["Date"])
df_change = pd.DataFrame(df_merged["BTC"].pct_change())
for s in stocklist:
    df_change = pd.concat([df_change, df_merged[s].pct_change()], axis=1)
```

```python
%matplotlib notebook
df_change[0:390].plot(figsize=(8, 10)).axhline(color='black', linewidth=2)
```

```python
change_sum = df_change.sum()*100
```

```python
change_sum
```

# PART 5 - DAILY LOG RETURN 

```python
df_merged = df_merged.sort_values("Date")
```

```python
df_log = pd.DataFrame(np.log(df_merged["BTC"] /df_merged["BTC"].shift(1)))
for s in stocklist:
    df_log = pd.concat([df_log, np.log(df_merged[s] /df_merged[s].shift())], axis=1)
```

```python
df_log = df_log.fillna(0)
df_log
```

```python
fig,ax=plt.subplots()
df_log.plot(ax=ax)
```

```python
log_sum = df_log.sum()*100
```

```python
log_sum
```

# PART 6 - VOLATILITY (ANNUALIZED)

```python
coin_list = ['BTC','ETH','ADA', 'BNB','XRP','SOL','DOGE','DOT','UNI', 'LINK']
volatility=[]
for s in coin_list:
      volatility.append(str(((df_log[s].std()*(np.sqrt(365))).round(2))*100))
dict_vol = dict(zip(coin_list,volatility))    
dict_vol
```

```python
fig,ax = plt.subplots()
df_change["BTC"].hist(ax=ax,bins=200, alpha=1, color = "orange")
ax.set_xlabel ("%change")
ax.set_ylabel("Frequency of Log Return")
ax.set_title("BTC VOLATILTY : "+ dict_vol["BTC"]+"%")
```

```python
fig,ax = plt.subplots()
df_change["ETH"].hist(ax=ax,bins=200, alpha=1, color = "blue")
ax.set_xlabel ("%change")
ax.set_ylabel("Frequency of Log Return")
ax.set_title("ETH VOLATILTY: "+ dict_vol["ETH"]+"%")
```

# PART - 7 PUTTING EVERYTHING TOGETHER

```python
df_all = pd.DataFrame([dict_roi])
df_all = df_all.transpose()
df_all["%change"] = change_sum
df_all["%log_return"] = log_sum
df_all["%volatility"] = dict_vol.values()
df_all_list = ["roi", "per_change", "per_log_return","%Volatility"]
df_all.columns = df_all_list
```
